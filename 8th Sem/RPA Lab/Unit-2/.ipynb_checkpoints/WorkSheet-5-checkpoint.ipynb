```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_csv('student.csv')
x= dataset.iloc[:, 0:-1].values
y= dataset.iloc[:, -1].values
```

```python
from sklearn.model_selecetion import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandaredScaler
sc = StandaredScaler
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
ann = tf.keras.model.Sequential()
```

```python
ann.add(tf.keras.layers.Dense (utils=4, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense (utils=1, activation='sigmoid'))
```

```python
ann.comple(optimizer='adam',loss='binary_crossentropy',matrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size=50, epochs=20)
```

```python
y_pred=ann.pridict(X_test)
y_pred=(y_pred>0.5)
y_pred=y_pred.astype(int)
```

```python
drom sklear.matrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
accuracy_score(y_test,y_pred)
```
